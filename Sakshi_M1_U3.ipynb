```python
theta1=np.arange(360)
theta1_rad=np.deg2rad(theta1)
sintheta=np.sin(theta1_rad)
plt.plot(theta1_rad, sintheta)
plt.xlabel('Radians')
plt.ylabel('sin(x)')
plt.title("Sinusoidal curve")
plt.show()
```

```python
theta1=np.arange(360)
theta1_rad=np.deg2rad(theta1)
costheta=np.cos(theta1_rad)
plt.plot(theta1_rad, costheta)
plt.xlabel('Radians')
plt.ylabel('cos(x)')
plt.title("Cosine curve")
plt.show()
```

```python
sintheta=np.sin(theta1_rad)
costheta=np.cos(theta1_rad)
plt.figure(figsize=(10,5))
plt.plot(theta1_rad, sintheta)
plt.plot(theta1_rad, costheta)
plt.xlabel('Radians')
plt.ylabel('sin(x), cos(x)')
plt.legend(['Sine','Cosine'])
plt.annotate("Maxima",xy=(2,1))
plt.savefig("myplot.pdf")
plt.show()
```
